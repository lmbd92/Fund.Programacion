*Funcin lambda*

Una funcin lambda en Python es una funcin annima, es decir, una funcin que no tiene un nombre definido. Se utiliza para definir funciones simples que se utilizan una nica vez y que no necesitan ser definidas formalmente mediante la declaracin "def". Las funciones lambda se crean utilizando la palabra clave "lambda" seguida de los parmetros de entrada separados por comas, seguidos por una expresin que define el resultado de la funcin.



```python
suma = lambda a,b: a + b
res = suma(3,2)
print(res)
```

```python
times_two= lambda var: var * 2
valor= times_two(10)
print(valor)
```

```python
class LightSwitch():

    def __init__(self):
        self.switchOn = 0

    def turnOn(self):
        self.switchOn = 1

    def turnOff(self):
        self.switchOn = 0

    def show(self):
        print("Switch is", "on" if self.switchOn else "off")


#Llamado de la clase
lightSwitch = LightSwitch()

lightSwitch.show()
lightSwitch.turnOn()
lightSwitch.show()
```

```python
class Persona:
    def __init__(self, nombre, sexo, profesion):
        self.nombre = nombre
        self.sexo = sexo
        self.profesion = profesion
    
    def show(self):
        print(f'Nombre: {self.nombre} Sexo: {self.sexo} Profesin: {self.profesion}')

    def Trabajo(self):
        print(f"{self.nombre} trabaja como {self.profesion}")
```

```python
# Ejemplo 1
p1 = Persona("Juan", "masculino", "programador")
p1.Trabajo()

# Ejemplo 2
p2 = Persona("Mara", "femenino", "mdico")
p2.Trabajo()

# Ejemplo 3
p3 = Persona("Pedro", "masculino", "abogado")
p3.Trabajo()
```
