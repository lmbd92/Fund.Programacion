```python
#Funcion que cuente el numero de caracteres = ("#")
def count_signo(st):
  count = 0
  for word in st.lower().split():
    if word == "#":
      count += 1
  return count

count_signo("# Impredecibles # Amigos # videojuegos")
```

```python
class LightSwitch():
  def __init__(self):
    self.switchOn = False

  def turnOn(self):
    self.switchOn = True

  def turnOff(self):
    self.switchOn = False

  def show(self):
    print(self.switchOn)

#llamamos la clase

LightSwitch = LightSwitch()

LightSwitch.show()
LightSwitch.turnOn()
LightSwitch.show()
LightSwitch.turnOff()
LightSwitch.show()
LightSwitch.turnOn()
LightSwitch.show()
```

```python
class Persona:
  def __init__(self, nombre, sexo, profesion):
    #instanciar las variables
    self.nombre = nombre
    self.sexo = sexo
    self.profesion = profesion

  #instanciar los metodos
  def show(self):
    print("Nombre : ", self.nombre, "sexo: ", self.sexo, "profesion: ", self.profesion)
  
  def work(self):
    print(self.nombre , "trabaja como ", self.profesion)

#creamos el objeto de tipo persona
tatiana = Persona("Tatiana", "Femenino", "Desarrollador")
pedro = Persona("Pedro", "Masculino", "Ingeniero")

#llamar los metodos de ese objeto
tatiana.show()
tatiana.work()
pedro.work()
```

```python
#herencia
class Vehicle:
  def __init__(self, name, color, price):
    self.name = name
    self.color = color
    self.price = price

  def show(self):
    print('Details: ', self.name, self.color, self.price)

  def max_speed(self):
    print("Vechicle max speed is 250")

  def engine(self):
    print("2500 HP")

# Clase Hija que va a heredar de la clase padre 
class Car(Vehicle):
  def max_speed(self):
    print("Car max spedd is 150")

  def engine(self):
    print("100 HP")

  #Creamos objeto de tipo Vehiculo => carro
car = Car('Car Mazda x3', 'Red', 100000 )
car.show()
car.max_speed()
car.engine()
#Creamos objeto de tipo vechiculo
vehicle = Vehicle('Tractomula x2', 'azul', 670000000)
vehicle.show()
vehicle.max_speed()
vehicle.engine()


```

```python
#Poliformismo

class Ferrari:
  def fuel_type(self):
    print("Petrol")

  def max_speed(self):
    print("Max speed 350km/h")

class BMW:
  def fuel_type(self):
    print("Diesel")

  def max_speed(self):
    print("Max speed is 240")

ferrari = Ferrari()
bmw = BMW()

#iterar objetos del mismo tipo

for car in (ferrari,bmw):
  car.fuel_type()  #llamamos los metodos sin revisar la clase del objeto
  car.max_speed()
```

Cuantas horas trabaja un empleado y cual es su salario de acuerdo a las horas trabajadas

```python
class Employee:
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

  def __mul__(self, timesheet):
    print(" Work for ", timesheet.days, 'days')
    #calculamos el salario
    return self.salary * timesheet.days

class TimeSheet:
  def __init__(self, name, days):
    self.name = name
    self.days = days

emp = Employee("Tatiana", 200)
timesheet = TimeSheet("Tatiana", 10)
print("El salario es : ", emp*timesheet)
```

```python
#herencia
# En una tienda se venden productos tipo: Adornos, alimentos, libros
#Referencia : Nombre, precio , Descripción
# si es alimento tiene un productor y un distribuidor
# si es libro tiene ISBN , Autor
```

```python
#Super clases

class Producto:
  def __init__(self, referencia, nombre, pvp, descripcion):
    self.referencia = referencia
    self.nombre = nombre
    self.pvp = pvp
    self.descripcion = descripcion

  def __str__(self):
    return f"REFERENCIA \t {self.referencia} \n" \
    f"NOMBRE\t \t {self.nombre}\n" \
    f"PRECIO\t \t{self.pvp} \n" \
    f"DESCRIPCION\t {self.descripcion}" 

class Adorno(Producto):
  pass


adorno = Adorno(2034, "Vaso", 1500, "Vaso de porcelana escarchado")
print(adorno)
          
```

```python
class Alimento(Producto):
  productor = ""
  distribuidor = ""

  def __str__(self):
    return f"REFERENCIA \t {self.referencia} \n" \
    f"NOMBRE\t \t {self.nombre}\n" \
    f"PRECIO\t \t{self.pvp} \n" \
    f"DESCRIPCION\t {self.descripcion}\n" \
    f"PRODUCTOR\t {self.productor}\n" \
    f"DISTRIBUIDOR\t {self.distribuidor}" \


al = Alimento(2000, "Botella Aceite", 9000, "Botella de 250 mL")
al.productor = "Oliosoya"
al.distribuidor = "Distriya"
print(al)
```

```python
class Libro(Producto):
  isbn = ""
  autor = ""

  def __str__(self):
    return f"REFERENCIA \t {self.referencia} \n" \
    f"NOMBRE\t \t {self.nombre}\n" \
    f"PRECIO\t \t{self.pvp} \n" \
    f"DESCRIPCION\t {self.descripcion}\n" \
    f"ISN\t\t {self.isbn}\n" \
    f"AUTOR\t\t {self.autor}" 

  
libro = Libro(2456, "100 años de soledad", 90500, "Libro colombiano")
libro.isbn = "0-1234-56-7"
libro.autor = "Gabriel Garcia Marquez"
print(libro)
```

```python
#Sumar y restar tiempo con timedelta
from datetime import datetime, timedelta

dt = datetime.now()
print(dt.strftime("%A %d de %B del %Y - %H:%M"))

t = timedelta(days = 14, hours = 4, seconds = 1000)

dentro_de_dos_semanas = dt + t
print(dentro_de_dos_semanas.strftime("%A %d de %B del %Y - %H:%M"))

hace_dos_semanas = dt - t
print(hace_dos_semanas.strftime("%A %d de %B del %Y - %H:%M"))
```

FICHERO CSV

----
Comad Separated Values (valores separados por comas)

```python
#Escritura de listas en csv

import csv

contactos = [
    ("Manuel", "Desarrollador", "manuel@ejemplo.com"),
    ("Lorena", "Gestora", "lorena@ejemplo.com"),
    ("Javier", "Analista", "javier@ejemplo.com"),
    ("Martha", "Directora", "martha@ejemplo.com")
]


with open("contactos.csv", "w", newline="\n") as csvfile:
  campos = ["nombre", "empleo", "email"]
  writer = csv.DictWriter(csvfile, delimiter = ',', fieldnames = campos)
  writer.writeheader()
  for nombre , empleo, email in contactos:
    writer.writerow({
        "nombre" : nombre, "empleo" : empleo, "email": email
    })
```

```python
with open("contactos.csv", newline="\n") as csvfile:
    reader = csv.DictReader(csvfile)
    for contacto in reader:
        print(contacto["nombre"], contacto["empleo"], contacto["email"])
```

```python
#Escritura de datos en JSON

import json

contactos = [
    ("Manuel", "Desarrollador Web", "manuel@ejemplo.com"),
    ("Lorena", "Gestora de proyectos", "lorena@ejemplo.com"),
    ("Javier", "Analista de datos", "javier@ejemplo.com"),
    ("Marta", "Experta en Python", "marta@ejemplo.com")
]

datos = []

for nombre, empleo, email in contactos:
  datos.append({"nombre ":nombre, "empleo": empleo, "email": email})

with open("contactos.json", "w") as jsonfile:
  json.dump(datos, jsonfile)
```

```python
#Lectura de datos en json

with open("contactos.json") as jsonfile:
  datos = json.load(jsonfile)
  for contacto in datos:
    print(contacto["nombre "], contacto["empleo"], contacto["email"])
```
